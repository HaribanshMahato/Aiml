```python
# Copy original DataFrame to avoid modifying inplace
df_filled = df.copy()

# Fill categorical columns with mode (most frequent value)
categorical_cols = ["EthnicGroup", "ParentEduc", "TestPrep", "ParentMaritalStatus",
                    "PracticeSport", "IsFirstChild", "TransportMeans", "WklyStudyHours"]

for col in categorical_cols:
    mode_value = df[col].mode()[0]  # Get mode without modifying df
    df_filled[col] = df[col].fillna(mode_value)
    print(f"Mode for {col}: {mode_value}")  # Print mode for reference

# Fill numerical column with median (to handle skewed data)
df_filled["NrSiblings"] = df["NrSiblings"].fillna(df["NrSiblings"].median())

# Check if missing values are filled
print(df_filled.isnull().sum())
```

```python
# Replace MathScore < 40 with NaN
df["MathScore"] = df["MathScore"].mask(df["MathScore"] < 40, float("nan"))
df
```

# grouping and agg

```python
# Group by ParentEduc and get the average MathScore for each education level
df.groupby("ParentEduc")["MathScore"].mean()
```

```python
# Group by ParentEduc and get the average MathScore for each education level
df.groupby("ParentEduc")["MathScore"].agg(['sum','mean','count'])
```

```python
df["NormalizedMath"] = df.groupby("ParentEduc")["MathScore"].transform(lambda x: (x - x.mean()) / x.std())
```
